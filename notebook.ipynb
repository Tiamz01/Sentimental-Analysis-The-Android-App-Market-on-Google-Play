```python
# Use this cell to begin your analysis, and add as many as you would like!
import pandas as pd

apps = pd.read_csv('datasets/apps.csv')

# Data Cleaning
chars_to_replace = ['+', ',']
    
for char in chars_to_replace:
    apps['Installs'] = apps['Installs'].apply(lambda x: x.replace(char, ''))

# converting to interger
apps['Installs'] = apps['Installs'].astype(int)


```

```python
# Groupin by app category and applying aggregate
app_category_info = apps.groupby('Category').agg({'App':'count', 'Price':'mean', 'Rating':'mean'})

# Renaming each column
app_category_info = app_category_info.rename(columns = {'App':'Number of apps', 'Price': 'Average price','Rating':'Average rating'})

```

```python

# Searching free finance app
free_finance_app = apps[(apps['Category'] == 'FINANCE' ) & (apps['Type'] == 'Free')]


# Reading the review file
reviews = pd.read_csv('datasets/user_reviews.csv')

# join the dataframe

merge_df = pd.merge(free_finance_app, reviews,  on= "App", how='inner')

# finding the average Sentiment
app_sentimental_score = merge_df.groupby('App').agg({'Sentiment Score': 'mean'})

# finding the sort
sort_app_sentimental_score  = app_sentimental_score.sort_values(by='Sentiment Score', ascending=False)

top_10_user_feedback = sort_app_sentimental_score[:10]

top_10_user_feedback
```
